```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
req = Request(
    url='https://remoteok.com/', 
    headers={'User-Agent': 'Mozilla/5.0'}
)
page = urlopen(req).read()
```

```python
dados = []
```

```python
soup = BeautifulSoup(page, "html.parser")
table = soup.find("table", {'id': 'jobsboard'})

rows =  table.find_all("tr", {'class': 'job'})

for row in rows:
    company = row.get('data-company')
    job_title = row.find("h2", {"itemprop":"title"}).text.strip()
    job_url = "https://remoteok.io" + row.find("a", href=True)["href"]
    
    dados.append([company, job_title, job_url])
```

```python
df = pd.DataFrame(dados, columns=["Empresa", "Posição", "URL"])
```
